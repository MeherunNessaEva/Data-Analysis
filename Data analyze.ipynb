```python
ax=sns.countplot(x="Gender",data=df)
#show the graph of gender 
```

```python
ax=sns.countplot(x="Gender",data=df)
for bars in ax.containers:
 ax.bar_label(bars)
#show the graph of gender with level bar
```

```python
df.groupby(["Gender"], as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
#show the sum of amount for each gender
```

```python
sales_gen=df.groupby(["Gender"], as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.barplot(x="Gender",y="Amount",data=sales_gen)
#show the sum of amount for each gender in bar chart
```

```python
sns.countplot(data=df,x="Age Group",hue='Gender')
#this shows different ages for each gender
```

```python
ax=sns.countplot(data=df,x="Age Group",hue='Gender')
for bars in ax.containers:
 ax.bar_label(bars)
#this shows different ages for each gender with level bar
```

```python
#shows datafram of used total amount for different age group 
df.groupby(["Age Group"], as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
```

```python
sales_age=df.groupby(["Age Group"], as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.barplot(x="Age Group",y="Amount",data= sales_age)
#shows bar chart of used total amount for different age group 
```

```python
df.columns
#show the attribute of columns level
```

```python
sales_state=df.groupby(["State"], as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=False)
sns.set(rc={"figure.figsize":(30,5)})
sns.barplot(x="State",y="Orders",data= sales_state)
#shows bar chart of total order for different states 
```

```python
sales_state=df.groupby(["State"], as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.set(rc={"figure.figsize":(30,5)})
sns.barplot(x="State",y="Amount",data= sales_state)
#shows bar chart of used total amount for different states 
```

```python
ax=sns.countplot(x="Marital_Status",data=df)
sns.set(rc={"figure.figsize":(10,5)})
for bars in ax.containers:
 ax.bar_label(bars)
#show the graph of each marital status with level bar
```

```python
sales_state=df.groupby(["Marital_Status","Gender"] ,as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.set(rc={"figure.figsize":(10,5)})
sns.barplot(x="Marital_Status",y="Amount",hue ="Gender",data= sales_state)
#show the sum of amount for each marital status in bar chart
```

```python
sns.set(rc={"figure.figsize":(30,5)})
ax=sns.countplot(x="Occupation",data=df)
for bars in ax.containers:
 ax.bar_label(bars)
 #number of orders , occupationwise with label bar
```

```python
sales_state=df.groupby(["Occupation"] ,as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.set(rc={"figure.figsize":(20,5)})
sns.barplot(x="Occupation",y="Amount",data= sales_state)
#show the sum of amount for each type occupation in bar chart
```

```python
#shows number of product catagory with label bar
sns.set(rc={"figure.figsize":(25,5)})
ax=sns.countplot(x="Product_Category",data=df)
for bars in ax.containers:
 ax.bar_label(bars)
```

```python
sales_state=df.groupby(["Product_Category"] ,as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.set(rc={"figure.figsize":(25,5)})
sns.barplot(x="Product_Category",y="Amount",data= sales_state)
#show the sum of amount for each type product in bar chart
```

```python
sales_state=df.groupby(["Product_ID"] ,as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=False).head()
sns.set(rc={"figure.figsize":(25,5)})
sns.barplot(x="Product_ID",y="Orders",data= sales_state)
 #number of orders , product_Id wise 
```
